## Analisar o valor de venda em escala nacional: 2022 - 1

```python
media_nacional = base2022_01['Valor de Venda'].mean()   #Cálculo do valor de venda médio
media_nacional = round(media_nacional,2)   #Arredondamento do valor de venda médio até os centavos
print("Média: R$",media_nacional,'\n') #Exibição em tela

mediana_nacional = base2022_01['Valor de Venda'].median()   #Cálculo da mediana do valor de venda
precoMedioNacional = round(mediana_nacional,2)   #Arredondamento do valor de venda médio até os centavos
print("Mediana: R$",mediana_nacional,'\n')   #Exibição em tela

moda_nacional = statistics.mode(base2022_01['Valor de Venda'])   #Cálculo da moda do valor de venda
moda_nacional = round(moda_nacional,2)   #Arredondamento do valor de venda médio até os centavos
print("Moda: R$",moda_nacional, '\n')   #Exibição em tela

variancia_nacional = base2022_01['Valor de Venda'].var()   #Cálculo da variância do valor de venda
variancia_nacional = round(variancia_nacional,2)   #Arredondamento da variância até os centavos
print("Variância: R$",variancia_nacional, '\n')   #Exibição em tela
        
desvio_padrao_nacional = base2022_01['Valor de Venda'].std()   #Cálculo do desvio padrão do valor de venda
desvio_padrao_nacional = round(desvio_padrao_nacional,2)   #Arredondamento do desvio padrão até os centavos
print("Desvio padrão: R$",desvio_padrao_nacional)   #Exibição em tela
```

## Criar função para plotar

```python
def seleciona_intervalo (base2022_01,produto):
    tabela = base2022_01.loc[base2022_01['Produto'] == str(produto)] 
    x = tabela['Data da Coleta'].to_list()
    y = tabela['Valor de Venda']
  
    plt.plot(x,y)
```

## Analisar o valor de venda em escala nacional: 2004 a 2022

```python
lista_salario_minimo = [260,260,300,300,350,350,380,380,415,415,465,465,510,510,540,545,622,622,678,
                        678,724,724,788,788,880,880,937,937,954,954,998,998,1039,1045,1100,1100,1212]

lista_bases = [base2004_01, base2004_02, base2005_01, base2005_02, base2006_01, base2006_02, base2007_01, base2007_02,
               base2008_01, base2008_02, base2009_01, base2009_02, base2010_01, base2010_02, base2011_01, base2011_02,
               base2012_01, base2012_02, base2013_01, base2013_02, base2014_01, base2014_02, base2015_01, base2015_02,
               base2016_01, base2016_02, base2017_01, base2017_02, base2018_01, base2018_02, base2019_01, base2019_02,
               base2020_01, base2020_02, base2021_01, base2021_02, base2022_01]

lista_datas = ['04-1','04-2','05-1','05-2','06-1','06-2','07-1','07-2','08-1','08-2','09-1','09-2',
               '10-1','10-2','11-1','11-2','12-1','12-2','13-1','13-2','14-1','14-2','15-1','15-2',
               '16-1','16-2','17-1','17-2','18-1','18-2','19-1','19-2','20-1','20-2','21-1','21-2','22-1']
```

```python
media_nacional_geral = base_geral['Valor de Venda'].mean()   #Cálculo do valor de venda médio
media_nacional_geral = round(media_nacional_geral,2)   #Arredondamento do valor de venda médio até os centavos
print("Média: R$",media_nacional_geral,'\n') #Exibição em tela

mediana_nacional_geral = base_geral['Valor de Venda'].median()   #Cálculo da mediana do valor de venda
precoMedioNacional_geral = round(mediana_nacional_geral,2)   #Arredondamento do valor de venda médio até os centavos
print("Mediana: R$",mediana_nacional_geral,'\n')   #Exibição em tela

moda_nacional_geral = statistics.mode(base_geral['Valor de Venda'])   #Cálculo da moda do valor de venda
moda_nacional_geral = round(moda_nacional_geral,2)   #Arredondamento do valor de venda médio até os centavos
print("Moda: R$",moda_nacional_geral, '\n')   #Exibição em tela

variancia_nacional_geral = base_geral['Valor de Venda'].var()   #Cálculo da variância do valor de venda
variancia_nacional_geral = round(variancia_nacional_geral,2)   #Arredondamento da variância até os centavos
print("Variância: R$",variancia_nacional_geral, '\n')   #Exibição em tela
        
desvio_padrao_nacional_geral = base_geral['Valor de Venda'].std()   #Cálculo do desvio padrão do valor de venda
desvio_padrao_nacional_geral = round(desvio_padrao_nacional_geral,2)   #Arredondamento do desvio padrão até os centavos
print("Desvio padrão: R$",desvio_padrao_nacional_geral, '\n')   #Exibição em tela

#Cálculo da média ponderada, para considerar a variação do salário mínimo ao longo do tempo
lista_media_preco_nacional_por_semestre = []
soma_pesos = 0
soma_numerador = 0

for base in lista_bases:
    lista_media_preco_nacional_por_semestre.append(base['Valor de Venda'].mean()) #Armazena o valor de venda médio no semestre

for salario_minimo in lista_salario_minimo:
    soma_pesos = soma_pesos + 1/salario_minimo # Calcula o denominador da média ponderada

for pos in range(0, len(lista_salario_minimo)):     
    soma_numerador = soma_numerador + lista_media_preco_nacional_por_semestre[pos]/lista_salario_minimo[pos] 
    #Calcula o númerador da média ponderada
    
media_ponderada = soma_numerador/soma_pesos
media_ponderada = round(media_ponderada,2)
print("Média ponderada: R$",media_ponderada, '(considerando o salário mínimo de cada período)\n')   #Exibição em tela
```

## Plotar gráfico Valor de Venda x Data da Coleta - Escala nacional: 2004 a 2022

```python
lista_porcentagem = []

#Relaciona o valor de venda médio no semestre e o salário mínimo naquele período

for bases in lista_bases:
    
    for lista in lista_salario_minimo:
        divisao_por_salario = bases['Valor de Venda'].div(lista) # Tabela com razão entre o valor e salário mínimo do período
        media_nacional_por_salario = divisao_por_salario.mean() # Calcula a média geral
        media_nacional_por_salario = round(media_nacional_por_salario*100,2) # Formata em porcentagem com precisão de 2 casas 
                                                                             #decimais
    lista_porcentagem.append(media_nacional_por_salario)  # Armazena os valores em uma lista
```

```python
x = lista_datas
y = lista_porcentagem 
plt.figure(figsize =(22,9), facecolor = 'white') #Dimensão e cor de fundo
plt.xlabel('Data da coleta') #Nome do eixo X
plt.ylabel('Valor do combustível por salário mínimo (%)') #Nome do eixo Y
plt.title('Análise do valor de venda nacional 2004 a 2022',fontweight = 'bold')
plt.plot(x,y,'k')
plt.grid(True)
plt.show
plt.savefig('Graficos\\Nacional\\Análise do valor de venda nacional 2004 - 2022.png')
```
